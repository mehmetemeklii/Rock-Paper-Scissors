```python
#Taş-Kağıt-Makas oyununu ingilizce olarak oynanmalıdır.
#1- Bilgisayar olmalıdır 
#2- Oyuncu sayısı girilmelidir
#3- Oyuncunun bilgisayar mı oyuncu mu olacağı seçilmeli
#4- Round sayısı girilmeli
#5- Oyuncu girilirse oyuncunun seçtiği olay, eğer bilgisayar girilirse random atama sağlanmalıdır
```

```python
#Random library run
import random


#Create action_list Paper, Rock, Scissors
action_list=["Paper","Rock","Scissors"]
loop_control = 0
#Ask the user how many people want to play
while loop_control == 0 :
  
  try:
    print('\nIF YOU EXIT THE APPLICATION, YOU ENTER 50 VALUE IN COMPUTER-PLAYER-TOTAL \n')

    computer_number = int(input('How many computer want to play? Please enter a number:'))
    if int(computer_number) < 0:
      print('\nENTERED NOT INTEGER. PLEASE ENTER INTEGER!\n')
    elif int(computer_number) == 50:
      loop_control = 1 
      continue   
    
    player_number = int(input('How many players want to play? Please enter a number:'))
    if int(player_number) < 0:
      print('\nENTERED NOT INTEGER. PLEASE ENTER INTEGER!\n')
    elif int(player_number) == 50:
      loop_control = 1
      continue
    
    total_round = int(input('How many rounds want to play? Please enter a number:'))
    if int(total_round)  < 0:
      print('\nENTERED NOT INTEGER. PLEASE ENTER INTEGER!\n')
    elif int(total_round) == 50:
      loop_control = 1
      continue

  except ValueError:
    print('\nYOU NOT A NUMBER YOUR ENTERED. PLEASE YOU ENTER A NUMBER!\n')


  else:
    computer_action_list, players_action_list,computer_score_list,player_score_list = {},{},{},{}
    round_counter = 0 
    x = 1
    i = 1
    while int(i)<= int(computer_number):
      computer_action_list[i]=random.choice(action_list)
      computer_score_list[i]=0
      i += 1
      if int(i) > int(computer_number):
        break

    while True:    
      round_counter += 1
      print('\nRound:',round_counter)
      print('\n')
      while int(x) <= int(player_number):
        try:
          players_action_list[x]=input(f'Please players {x} enter action:').capitalize()
          player_score_list[x]=0
          if players_action_list[x]=='Paper' or players_action_list[x]=='Rock' or players_action_list[x]=='Scissors' :
            x += 1
          else:
            print('\nYOUE ENTERED WRONG VALUE. PLEASE ROCK-PAPER-SCISSORS ENTER!\n')
          continue
          
        except:
          if int(x) > int(player_number):
            x = 1
            break

      print('\nComputers choice list = ' ,computer_action_list)
      print('Players choice list = ' ,players_action_list)
      print('\n\t\tMATCH LIST:\n')

      for computer_items in computer_action_list.items():
        for players_items in players_action_list.items():
          a,computer_choice = computer_items
          b,player_choice = players_items
          if computer_choice == player_choice:
            print(f'Tie! Both players choise the smae action. The same players Computer {a} - Player {b}')
          elif computer_choice == 'Paper':
            if player_choice == 'Rock':
              print(f'Played Computer {a} wiht Player {b}. Winner is: Computer {a}')
              computer_score_list[a] += 1
            elif player_choice == 'Scissors':
              print(f'Played Computer {a} wiht Player {b}. Winner is: Player {b}')
              player_score_list[b] += 1
          elif computer_choice == 'Rock':
            if player_choice == 'Paper':
              print(f'Played Computer {a} wiht Player {b}. Winner is: Player {b}')
              player_score_list[b] += 1
            elif player_choice == 'Scissors':
              print(f'Played Computer {a} wiht Player {b}. Winner is: Computer {a}')
              computer_score_list[a] += 1
          elif computer_choice == 'Scissors':
            if player_choice == 'Paper':
              print(f'Played Computer {a} wiht Player {b}. Winner is: Computer {a}')
              computer_score_list[a] += 1
            elif player_choice == 'Rock':
              print(f'Played Computer {a} wiht Player {b}. Winner is: Player {b}')
              player_score_list[b] += 1

      if total_round == round_counter:
        break
    print('\n\t\tTOTAL SCORE TABLE LIST:')
    for i in computer_score_list.items():
      for x in player_score_list.items():
        c_keys,c_values = i
        p_keys,p_values = x
        if int(c_values) == int(p_values):
          print(f'\nThere is no winner, tie.\t Computer {c_keys} - Player {p_keys}. \t Score = {computer_score_list[c_keys]} :  {player_score_list[p_keys]} ')
        elif int(c_values) > int(p_values):
          print(f'\nComputer {c_keys} won.\t Computer {c_keys} - Player {p_keys}.\t \t Score =  {computer_score_list[c_keys]} :  {player_score_list[p_keys]} ')
        elif int(c_values) < int(p_values):
          print(f'\nPlayer {p_keys} won.\t Computer {c_keys} - Player {p_keys}. \t \t Score =  {computer_score_list[c_keys]} :  {player_score_list[p_keys]} ')
```
